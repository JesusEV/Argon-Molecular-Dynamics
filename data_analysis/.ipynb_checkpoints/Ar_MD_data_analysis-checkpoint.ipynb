```python
from tqdm import tqdm
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
N = 108
L =34.1580
V = L**3
rho = N/L
sigma = 3.4
radii = np.linspace(0,30,100, endpoint=True)
```

```python
file_name = 'distances.dat'
distances =  np.genfromtxt(file_name, delimiter=" ", usecols=range(0,1), unpack=False)
distances = distances[distances != 0]
histogram_distances = np.histogram(distances,radii, density=False)[0]
```

```python
plt.title('Radial Distance Histogram')
n3, bins3, patches3 = plt.hist(distances, 100, density=True, facecolor='g', alpha=0.75, label='Generated Histogram')
plt.grid(True)
plt.legend()
plt.xlabel('r')
plt.ylabel('f(r)')
plt.show()
```

```python
dr = radii[1]

RDF = histogram_distances
RDF = RDF/rho
for i, r in enumerate(radii[1:]):
    dr = radii[i+1] - radii[i]
    RDF[i] = RDF[i]/(4 * np.pi * r**2 * dr)
```

```python
plt.title('Radial Distribution Functions')
plt.plot(radii[1:]/sigma, RDF, label='Temp={T}')
plt.grid(True)
plt.legend()
plt.xlabel('r')
plt.ylabel('RDF(r)')
plt.xlim(0,4)
plt.show()
```
